## Players Data Analysis

In this notebook, we’ll explore the player_data dataset and player_financial datasets and merge them.


### Loading the Dataset
We will load in all our packages and our datasets all at once in the beginning.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

baby = pd.read_csv('players_data-2024_2025.csv')
toddler = pd.read_csv('Player_Financial.csv')
```

### Merging the Datasets
Now, we will merge our two datasets into one.

```python

merged_data = pd.merge(baby, toddler, on="Player", how="inner")

merged_data.to_csv("merged_player_data.csv", index=False)

merged_data.head()
```

### Why I chose Inner Merge

I opted for an inner join to ensure that only records appearing in both datasets are included in the final merged dataset. This approach:

Eliminates unmatched records, preserving data accuracy.
Ensures completeness, as every Player in the final dataset contains both Player data and Financial Data.
